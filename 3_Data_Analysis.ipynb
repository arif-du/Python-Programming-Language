**What is Plate Carree?**

Plate Carrée (French for "flat square") is one of the simplest map projections.

It directly maps latitude and longitude to a rectangular grid:

X-axis = Longitude

Y-axis = Latitude

It's commonly used for quick plotting because it's easy to understand and apply, though it distorts areas near the poles

image.png

Create a 10x5 inch figure

Use a simple latitude/longitude projection

Draw coastlines

Focus on the Bay of Bengal region (80°E to 100°E, 5°N to 25°N)

Add labeled gridlines

Display the map

```python
# Import the PlateCarree projection system from Cartopy
import cartopy.crs as ccrs
# Import the plotting library
import matplotlib.pyplot as plt

# Create a figure with specific size (10 inches wide, 5 inches tall)
plt.figure(figsize=(10, 5))

# Create axes using the PlateCarree projection (a simple lat/lon grid projection)
ax = plt.axes(projection=ccrs.PlateCarree())

# Add coastlines to the map
ax.coastlines()

# Set the geographical boundaries: [min_lon, max_lon, min_lat, max_lat] (Bay of Bengal region)
ax.set_extent([80, 100, 5, 25])

# Add gridlines with labels for latitude and longitude
ax.gridlines(draw_labels=True)

# Add a title to the plot
plt.title("Base Map - Bay of Bengal")

# Display the map
plt.show()
```

```python
# Sample data (can be loaded from a CSV or dataset)
import pandas as pd

data = {
    'lat': [21.7, 21.8, 21.9, 22.0],
    'lon': [89.9, 90.0, 90.1, 90.2],
    'temp': [26.5, 27.0, 27.5, 28.0]  # temperature values
}

df = pd.DataFrame(data)
```

| Part               | Meaning                                                                   |
| ------------------ | ------------------------------------------------------------------------- |
| `plt.scatter(...)` | Creates a scatter plot (a set of dots on a plot).                         |
| `df['lon']`        | X-axis values: longitudes of the data points.                             |
| `df['lat']`        | Y-axis values: latitudes of the data points.                              |
| `c=df['temp']`     | Colors each point based on the corresponding **temperature value**.       |
| `cmap='coolwarm'`  | Color map: cool colors for low values, warm colors for high values.       |
| `s=100`            | Size of each point (in points²). Bigger value = bigger dots.              |
| `edgecolor='k'`    | Each dot has a **black ('k') border** for better visibility.              |
| `sc = ...`         | Stores the scatter object to be used later (e.g., in `plt.colorbar(sc)`). |



```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Create the map figure
plt.figure(figsize=(10, 5))
ax = plt.axes(projection=ccrs.PlateCarree())

# Add base map features
ax.coastlines()
ax.set_extent([89.8, 90.5, 21.6, 22.3])  # Kuakata region
ax.gridlines(draw_labels=True)

# Scatter plot with color mapping (temperature)
sc = plt.scatter(df['lon'], df['lat'], c=df['temp'], cmap='coolwarm', s=100, edgecolor='k')

# Add color bar
plt.colorbar(sc, label="Temperature (°C)")

# Title
plt.title("Temperature Mapping - Kuakata Region")

# Show map
plt.show()
```

# **GSW: (Gibbs SeaWater Oceanographic Package)**

GSW (Gibbs SeaWater Oceanographic Package) is a software library developed by the Thermodynamic Equation of Seawater – 2010 (TEOS-10) working group. It is the official toolbox recommended by the UNESCO/SCOR/IAPSO Joint Scientific Committee for oceanographic and climate studies.

```python
pip install gsw
```

**Calculate Absolute Salinity, Conservative Temperature, and Density**

```python
# Import the GSW (Gibbs SeaWater) oceanographic toolbox
import gsw

# --- Input parameters ---
SP = 35        # Practical Salinity (unitless)
t = 15         # In-situ temperature in degrees Celsius (°C)
p = 1000       # Pressure in decibar (dbar)
lon = 90       # Longitude in degrees East (°E)
lat = 21       # Latitude in degrees North (°N)

# --- Convert Practical Salinity to Absolute Salinity (g/kg) ---
# Required for accurate thermodynamic calculations
SA = gsw.SA_from_SP(SP, p, lon, lat)

# --- Convert in-situ temperature to Conservative Temperature (°C) ---
# Conservative Temperature is more suitable for heat content calculations in oceanography
CT = gsw.CT_from_t(SA, t, p)

# --- Calculate in-situ density of seawater (kg/m³) ---
# Based on Absolute Salinity and Conservative Temperature
rho = gsw.rho(SA, CT, p)

# --- Display the results ---
print(f"Absolute Salinity (g/kg): {SA:.2f}")         # e.g., 35.05
print(f"Conservative Temperature (°C): {CT:.2f}")    # e.g., 14.85
print(f"Density (kg/m³): {rho:.2f}")                 # e.g., 1027.25
```
